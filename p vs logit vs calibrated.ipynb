```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'p': np.linspace(0.00001, 0.99999, 100),
})
```

```python
from scipy.special import logit
df['logit'] = logit(df['p'])
```

```python
df['logp'] = np.log(df['p'])
```

```python
df.plot(x='p', y='logit')
```

```python
df.plot(x='p', y='logp')
```

```python
alpha = 0.1
beta = 1.1
```

```python
score = 'p'
df['calibrated'] = 1 / (1 + np.exp(alpha - beta * df[score]))
df.plot(x=score, y='calibrated')
```

```python
score = 'logit'
df['calibrated'] = 1 / (1 + np.exp(alpha - beta * df[score]))
df.plot(x=score, y='calibrated')
```

```python
score = 'logp'
df['calibrated'] = 1 / (1 + np.exp(alpha - beta * df[score]))
df.plot(x=score, y='calibrated')
```
